```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
base = os.path.split(os.getcwd())[0]
dd = base + '/data/'
df_all = pd.read_csv(dd + 'data_after_imputation.csv')
df_all.columns
```

```python
df_all
```

```python
df_hist= df_all.drop(['fips', 'SVI', 'CVAC', 'Urban_rural_code'], axis = 1)
df_hist.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8); 
plt.tight_layout()
```

```python
sns.pairplot(df_hist)
```

```python
df_hist[['pop_full_12plus_per',
       'pop_full_18plus_per', 'pop_full_65plus_per']].hist(figsize=(20, 20), bins=50, xlabelsize=10, ylabelsize=10); 
plt.tight_layout()
```

```python
sns.set(style="darkgrid")

fig, axs = plt.subplots(1, 3, figsize=(15, 6))
sns.histplot(data=df_hist, x='pop_full_12plus_per',kde=True,  ax=axs[0])
sns.histplot(data=df_hist, x='pop_full_18plus_per', kde=True,  ax=axs[1])
sns.histplot(data=df_hist, x='pop_full_65plus_per', kde=True,  ax=axs[2])
plt.show()
```

```python
sns.set(style="darkgrid")

fig, axs = plt.subplots(1, 2, figsize=(15, 6))
sns.histplot(data=df_hist, x='cases_prop',kde=True,  ax=axs[0])
sns.histplot(data=df_hist, x='death_prop', kde=True,  ax=axs[1])
plt.show()
```
